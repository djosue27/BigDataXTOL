```python
for col in credit.columns:
    if credit[col].dtype == "object":
        # If categorical, do a count bar plot
        lm = pd.value_counts(credit[col]).plot.bar()
        lm.set_xticklabels(lm.get_xticklabels(),rotation=0)
        plt.title(col)
        plt.show()

    else:
        # If numerical, do a histogram
        credit.hist(column=col, bins=100)
        plt.title(col)
        plt.show()
```
